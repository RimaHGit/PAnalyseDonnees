```python
import pandas as pd

import datetime
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt

# df1 = pd.read_csv(r"C:\Users\SAMSUNG\Desktop\OpenClassrooms\P2_OC\Projet2\P2_donnees\Edstats_csv\EdStatsData.csv")

indicators = ["Internet users (per 100 people)",
              "Enrolment in upper secondary education, both sexes (number)",
              "Enrolment in post-secondary non-tertiary education, both sexes (number)",
              "Enrolment in tertiary education, all programmes, both sexes (number)"]

dfc_nb0 = df1[df1["Indicator Name"].isin(indicators)]
for idx in range(len(indicators)):
    dfl_nb0 = dfc_nb0.loc[dfc_nb0['Indicator Name']==indicators[idx]]
    print(indicators[idx], len(dfl_nb0.index))
    lsty=[]
    lstx=[]
    minimum=999
    AnneeCorrespMin=1000
    for i in range(65): # par année
        nbValeursdispo = dfl_nb0.describe().iloc[0,i]
        nbValeursnulles = len(dfl_nb0.index) - nbValeursdispo # compte le nombre de valeurs nulles
        if minimum>nbValeursnulles: # recherche du nombre de valeurs nulles minimum
            minimum=nbValeursnulles
            AnneeCorrespMin=dfl_nb0.describe().columns[i]
        print("Nombre de valeurs nulles pour l'année", dfl_nb0.describe().columns[i], "=", nbValeursnulles)
        lstx.append(dfl_nb0.describe().columns[i])
        lsty.append(nbValeursnulles)
    print("Minimum pour cet indicateur: ", minimum, " année correspondante: ", AnneeCorrespMin)
    
    
    #histogram
    y = lsty
    x = lstx
    x_pos = [i for i, _ in enumerate(x)]
    plt.figure(figsize=(20,8))
    plt.bar(x_pos, y, color='blue')
    plt.xlabel("Année")
    plt.ylabel("Nombre")
    plt.title("Nombre de données manquantes par année "+indicators[idx])
    plt.xticks(x_pos, x, rotation=90)
    plt.show()
```

```python
# indicateurs quantitatifs :
# "Internet users (per 100 people)", 
# "Enrolment in upper secondary education, both sexes (number)", 


# Je calcule un indicateur : My Indicator 
# My Indicator=["Enrolment in upper secondary education, both sexes (number)"] * ["Internet users (per 100 people)"/100]
           


import pandas as pd

import datetime
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt

# df1 = pd.read_csv(r"C:\Users\SAMSUNG\Desktop\OpenClassrooms\P2_OC\Projet2\P2_donnees\Edstats_csv\EdStatsData.csv")


# afficher "Enrolment in upper secondary education, both sexes (number)" par pays
indicator1 = ["Enrolment in upper secondary education, both sexes (number)"]

eParPays = df1[df1["Indicator Name"].isin(indicator1)]
eParPays = eParPays[["Country Name","Indicator Name","2014"]].dropna() # on retire les NAN
eParPays = eParPays.groupby("Country Name")["2014"].sum()
eParPays # nombre d'inscrits au lycée


# afficher "Internet users (per 100 people)" par pays
indicator2 = ["Internet users (per 100 people)"]

iParPays = df1[df1["Indicator Name"].isin(indicator2)]
iParPays = iParPays[["Country Name","Indicator Name","2014"]].dropna() # on retire les NAN
iParPays = iParPays.groupby("Country Name")["2014"].sum()
iParPays # Internet users (per 100 people)

# nb inscrits connectés
lsty =[]
lstx =[]

for idx,valx in eParPays.iteritems():
    for idy,valy in iParPays.iteritems(): # double boucles pour comparer puis faire le produit 
        if idx == idy :                   # eParPays * iParPays
            lsty.append(idx)
            lstx.append(valx*valy/100) # valy = Internet users (per 100 people) = valy/100 = tx Internet users
           
        
# on fusionne les deux vecteurs dans un dataframe
df_3indic_EI = pd.DataFrame({'Country Name':lsty, 'My Indicator':lstx}) # my indicator = nb inscrits * tx Internet users

# on filtre 'Country Name' par pays 
dfpays_3indic_EI=df_3indic_EI[~df_3indic_EI['Country Name'].isin(['Arab World','East Asia & Pacific','East Asia & Pacific (excluding high income)','Euro area',
                                    'Europe & Central Asia','European Union','Europe & Central Asia (excluding high income)',
                                    'Heavily indebted poor countries (HIPC)','High income','Latin America & Caribbean',
                                    'Latin America & Caribbean (excluding high income)','Least developed countries: UN classification',
                                    'Low & middle income','Low & middle income','Low income','Lower middle income','Middle East & North Africa',
                                    'Upper middle income','Middle income','Middle East & North Africa (excluding high income)', 'North America',
                                    'OECD members','South Asia','Sub-Saharan Africa','Sub-Saharan Africa (excluding high income)','World'])]

dfpays_3indic_EI
```

```python
# indicateurs quantitatifs :
# "Internet users (per 100 people)", 
# "Enrolment in upper secondary education, both sexes (number)", 
# "Enrolment in post-secondary non-tertiary education, both sexes (number)", 
# "Enrolment in tertiary education, all programmes, both sexes (number)"


# Je calcule un indicateur : My Indicator 
# My Indicator=["Enrolment in upper secondary education, both sexes (number)" 
#           +"Enrolment in post-secondary non-tertiary education, both sexes (number)" 
#           +"Enrolment in tertiary education, all programmes, both sexes (number)"] * ["Internet users (per 100 people)"/100]
           


import pandas as pd

import datetime
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt

# df1 = pd.read_csv(r"C:\Users\SAMSUNG\Desktop\OpenClassrooms\P2_OC\Projet2\P2_donnees\Edstats_csv\EdStatsData.csv")

# total inscrits
indicators = ["Enrolment in upper secondary education, both sexes (number)", 
              "Enrolment in post-secondary non-tertiary education, both sexes (number)",
              "Enrolment in tertiary education, all programmes, both sexes (number)"]


df_te = df1[df1["Indicator Name"].isin(indicators)]
df_te = df_te[["Country Name","Indicator Name","2014"]].dropna() # on retire les NAN
totalParPays = df_te.groupby("Country Name")["2014"].sum()
totalParPays # nombre d'inscrits aux 3 niveaux par pays

# afficher "Internet users (per 100 people)" par pays
indicator = ["Internet users (per 100 people)"]

df_i = df1[df1["Indicator Name"].isin(indicator)]
iParPays = df_i[["Country Name","Indicator Name","2014"]].dropna() # on retire les NAN
iParPays = iParPays.groupby("Country Name")["2014"].sum()
iParPays # Internet users (per 100 people) par pays

# nb inscrits connectés
lsty =[]
lstx =[]

for idx,valx in totalParPays.iteritems():
    for idy,valy in iParPays.iteritems(): # double boucles pour comparer puis faire le produit 
        if idx == idy :                   # totalParPays * iParPays
            lsty.append(idx)
            lstx.append(valx*valy/100) # valy = Internet users (per 100 people)
                                       # valy/100 = tx Internet users
           
        
# on fusionne les deux vecteurs dans un dataframe
df_5indic_TEI = pd.DataFrame({'Country Name':lsty, 'My Indicator':lstx}) # my indicator = total inscrits * tx Internet users

# on filtre 'Country Name' par pays 
dfpays_5indic_TEI=df_5indic_TEI[~df_5indic_TEI['Country Name'].isin(['Arab World','East Asia & Pacific','East Asia & Pacific (excluding high income)','Euro area',
                                    'Europe & Central Asia','European Union','Europe & Central Asia (excluding high income)',
                                    'Heavily indebted poor countries (HIPC)','High income','Latin America & Caribbean',
                                    'Latin America & Caribbean (excluding high income)','Least developed countries: UN classification',
                                    'Low & middle income','Low & middle income','Low income','Lower middle income',
                                    'Middle East & North Africa','Upper middle income','Middle income',
                                    'Middle East & North Africa (excluding high income)', 'North America','OECD members','South Asia',
                                    'Sub-Saharan Africa','Sub-Saharan Africa (excluding high income)','World'])]

dfpays_5indic_TEI
```
